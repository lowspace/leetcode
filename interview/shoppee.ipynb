# problem 3

如果一个字符串仅由“A” “B”两个字符组成，且字符串内任意相邻两个字母不相同，我们称这个字符串为互斥字符串。

每一步操作中，我们可以将A变成B或者将B变成A，请问由一个仅由AB组成的字符串变成互斥字符串，最少需要多少步骤？


如果最新的一样：
+ 改自己 F(n) = F(n - 1) + 1
+ 改之前 F(n) = F(n - 2) + 1
  
如果最新的不一样:
F(n) = F(n - 1) or F(n - 2) + 1

```python
class Solution:
    def MinOperations(self, s) :
        # write code here
        pre, curr = 0, 0

        for i in range(1, len(s)):
            if s[i] == s[i - 1]:
                curr, pre = min(curr + 1, pre + 1), curr
            else:
                curr, pre = curr, curr

        return curr
```
