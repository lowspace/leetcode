# Problem

You are given a string `s` consisting **only** of letters `'a'` and `'b'`. In a single step you can remove one **palindromic subsequence** from `s`.

Return *the **minimum** number of steps to make the given string empty*.

A string is a **subsequence** of a given string if it is generated by deleting some characters of a given string without changing its order. Note that a subsequence does **not** necessarily need to be contiguous.

A string is called **palindrome** if is one that reads the same backward as well as forward.

 

**Example 1:**

```
Input: s = "ababa"
Output: 1
Explanation: s is already a palindrome, so its entirety can be removed in a single step.
```

**Example 2:**

```
Input: s = "abb"
Output: 2
Explanation: "abb" -> "bb" -> "". 
Remove palindromic subsequence "a" then "bb".
```

**Example 3:**

```
Input: s = "baabb"
Output: 2
Explanation: "baabb" -> "b" -> "". 
Remove palindromic subsequence "baab" then "b".
```

 

**Constraints:**

- `1 <= s.length <= 1000`
- `s[i]` is either `'a'` or `'b'`.

## [DEPRECATED] Brute Force

Since we need to return the minimum number to empty the string, thus for each step we may should remove the longest subsequence palindromic string.~~

Q1: Doest removing the longest subsequence palindromic string for each step guarantee the minimum operation time?

A1: Giving a situation, removing the first and second `a` in the string `[1]a[2]a[3]` obtains the long palindromic string `[1][2][3]`. Therefore, naive brute force can't guarantee the correctness, which removes the element one by one and checks the palindrome.

## Version 1 Brute Force

As the problem description said, the string consists of `a` and `b`, and the subsequence is not contiguous. Thus, we only need to check the palindrome of the whole string, otherwise, deleting all `a` and all `b` returns the empty string.

````python
class Solution:
    def removePalindromeSub(self, s: str) -> int:
        if s == s[::-1]:
            return 1
        else:
            return 2
        
````
